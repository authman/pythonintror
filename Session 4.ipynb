```python
class Student():
    def __init__(self, name, age, email):
        self.name = name
        self.age = age
        self.email = email
        
    def is_over_20(self):
        if self.age > 20:
            return 'Over 20'
            
        else:
            return 'Nope'
            
    def has_email(self):
        if self.email is None:
            return 'Nope'
            
        else:
            return 'Yes'
```

```python
a = Student('a',1,'a@b.com')
```

```python
answer = a.is_over_20()
print(answer)
```

```python
answer = a.has_email()
print(answer)
```
