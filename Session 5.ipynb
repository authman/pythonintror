```python
fig, ax = plt.subplots(1, 1, figsize=(5,5))

ax.set_title('Covid19 Survival')
ax.set_aspect('equal')
ax.set_xlim((0,BOUNDARY))
ax.set_ylim((0,BOUNDARY))

circles = ax.plot([], [], 'bo', ms=6)[0]

def init():
    """initialize animation"""
    # Set a random start position
    people[:, :2] = np.random.randint(
        0,
        BOUNDARY,
        size=(NUM_PEOPLE, 2)
    )

    # Set a random velocity
    people[:, 2:4] = np.random.randint(
        -3,
        3,
        size=(NUM_PEOPLE, 2)
    ) / 10

    # Infect the 0th-person
    people[0, 4] = 1 # 1=infected
    
    circles.set_data(
        people[:, 0],
        people[:, 1]
    )
    return circles,

def contiminate():
    for i in range(NUM_PEOPLE):
        if people[i,4] == 0:
            continue
            
        for j in range(NUM_PEOPLE):
            if i==j: continue
                
            dist = np.linalg.norm(
                people[i,:2] - people[j, 2:4]
            )
            
            if dist < 4:
                people[j, 4] = 1
                
                
def animate(i):
    global circles
    
    # Move a step in the direction of their velocity
    people[:, :2] += people[:, 2:4]
    
    
    # Check to see if they're outside of bounds
    mask = people[:,0] < 0
    people[mask, 0] = 0
    people[mask, 2] *= -1
    
    mask = people[:,1] < 0    
    people[mask, 1] = 0
    people[mask, 3] *= -1
    
    mask = people[:,0] > BOUNDARY
    people[mask, 0] = BOUNDARY
    people[mask, 2] *= -1
    
    mask = people[:,1] > BOUNDARY
    people[mask, 1] = BOUNDARY
    people[mask, 3] *= -1
    
    contiminate()
    
    circles.set_data(
        people[:, 0],
        people[:, 1]
    )

    return circles,


anim = animation.FuncAnimation(
    fig,
    animate,
    frames=1000,
    interval=10,
    blit=True, 
    init_func=init
)
HTML(anim.to_html5_video())
```
